```python
train = train[['nodal_skin_eruptions', 'chills', 'stomach_pain', 'vomiting', 'spotting_ urination', 'fatigue', 'weight_loss', 'high_fever', 'breathlessness', 'dark_urine', 'pain_behind_the_eyes', 'constipation', 'diarrhoea', 'mild_fever', 'yellowing_of_eyes', 'redness_of_eyes', 'chest_pain', 'fast_heart_rate', 'dizziness', 'extra_marital_contacts', 'slurred_speech', 'knee_pain', 'muscle_weakness', 'unsteadiness', 'bladder_discomfort', 'internal_itching', 'depression', 'muscle_pain', 'altered_sensorium', 'red_spots_over_body', 'abnormal_menstruation', 'increased_appetite', 'lack_of_concentration', 'receiving_blood_transfusion', 'stomach_bleeding', 'distention_of_abdomen', 'prominent_veins_on_calf', 'blackheads', 'small_dents_in_nails', 'blister','prognosis']].copy()
test = test[['nodal_skin_eruptions', 'chills', 'stomach_pain', 'vomiting', 'spotting_ urination', 'fatigue', 'weight_loss', 'high_fever', 'breathlessness', 'dark_urine', 'pain_behind_the_eyes', 'constipation', 'diarrhoea', 'mild_fever', 'yellowing_of_eyes', 'redness_of_eyes', 'chest_pain', 'fast_heart_rate', 'dizziness', 'extra_marital_contacts', 'slurred_speech', 'knee_pain', 'muscle_weakness', 'unsteadiness', 'bladder_discomfort', 'internal_itching', 'depression', 'muscle_pain', 'altered_sensorium', 'red_spots_over_body', 'abnormal_menstruation', 'increased_appetite', 'lack_of_concentration', 'receiving_blood_transfusion', 'stomach_bleeding', 'distention_of_abdomen', 'prominent_veins_on_calf', 'blackheads', 'small_dents_in_nails', 'blister','prognosis']].copy()
```

```python
X_train = train.drop(columns = ['prognosis'])
X_test = test.drop(columns = ['prognosis'])
y_train = train['prognosis']
y_test = test['prognosis']
```

```python
num_classes = len(set(y_train))
```

```python
label_encoder = LabelEncoder()
y_train_int = label_encoder.fit_transform(y_train)
y_train = to_categorical(y_train_int, num_classes=num_classes)
```

```python
model = models.Sequential()
model.add(layers.Dense(128, input_shape=(X_train.shape[1],), activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(64, activation='relu'))
model.add(layers.Dense(num_classes, activation='softmax'))
model.compile(optimizer='adam',
              loss='categorical_crossentropy',
              metrics=['accuracy'])

#y_train = to_categorical(y_train, num_classes=num_classes)
model.fit(X_train, y_train, epochs=10, batch_size=32, validation_split=0.2)
```

```python
y_test_int = label_encoder.transform(y_test)

# (Optional) One-hot encode y_test if needed
y_test_one_hot = to_categorical(y_test_int, num_classes=num_classes)
```

```python
loss, accuracy = model.evaluate(X_test, y_test_one_hot)

print(f"Test Loss: {loss}")
print(f"Test Accuracy: {accuracy}")
```

```python
# prompt: accuracy in percentages

accuracy_percentage = accuracy * 100
print(f"Test Accuracy: {accuracy_percentage:.2f}%")
```

```python
predictions = model.predict(X_test)
predicted_class_indices = np.argmax(predictions, axis=1)
predicted_labels = label_encoder.inverse_transform(predicted_class_indices)

# Compare with actual labels
actual_labels = label_encoder.inverse_transform(y_test_int)

print(f"Predicted Labels: {predicted_labels}")
print(f"Actual Labels: {actual_labels}")
```
